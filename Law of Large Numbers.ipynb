```python
from random import choice
```

### COIN FLIPPING

```python
coin = ['Heads', 'Tails']
```

```python
heads, tails = 0, 0

for i in range(100):
    result = choice(coin)
    
    if result == 'Heads' : heads += 1
    if result == 'Tails' : tails += 1
        
print("P(Heads) : ", 100 * heads / 100)
print("P(Tails) : ", 100 * tails / 100)
```
