Answer the following questions by calculating the number of ways of obtaining particular

#### <span style="color:purple">1. An athlete has eight different trophies, but only has room for four trophies in a display cabinet. How many different ways is it possible to display just four trophies out of eight, assuming that the display order is important?

```python
from itertools import permutations
```

```python
result = len(list(permutations([1,2,3,4,5,6,7,8], 4)))
result
```

```python
# second solution
```

```python
def fact(n):  
    return 1 if (n==1 or n==0) else n * fact(n - 1); 

def perm(n,r):
    return fact(n) / fact(n-r)

perm(8,4)
```

#### <span style="color:purple">2. A football manager has a squad of 20 players. How many different teams of 11 players could be selected from the squad? (Hint: Assume that positions of the players are not important)

```python
from itertools import combinations
```

```python
squad = list(range(1,21))

result = len(list(combinations(squad, 11)))
result
```

```python
# second solution
```

```python
def fact(n):  
    return 1 if (n==1 or n==0) else n * fact(n - 1); 

def comb(n,r):
    return fact(n) / (fact(n-r) * fact(r))

comb(20,11)
```
